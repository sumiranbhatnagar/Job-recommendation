****Recommender system based on the "jobtitle" and "experience" columns:****

```python
import os
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the naukri.com dataset
df = pd.read_csv('/kaggle/input/jobs-on-naukricom/naukri_com-job_sample.csv')

# Process the data
df['jobtitle'] = df['jobtitle'].apply(lambda x: x.lower())  # convert to lowercase
df['experience'] = df['experience'].str.extract('(\d+)').astype(float)  # extract years of experience
df = df.dropna(subset=['jobtitle', 'experience'])  # drop rows with missing values
df = df.drop_duplicates(subset=['jobtitle', 'experience'])  # drop duplicate rows

# Define the input variables
jobtitle_vectorizer = CountVectorizer()
jobtitle_matrix = jobtitle_vectorizer.fit_transform(df['jobtitle'])
experience_matrix = np.array(df['experience']).reshape(-1, 1)

# Build the recommender system
def get_similar_jobs(jobtitle, experience, top_n=10):
    jobtitle_vector = jobtitle_vectorizer.transform([jobtitle.lower()])
    similarity_scores = cosine_similarity(jobtitle_vector, jobtitle_matrix).flatten()
    similarity_scores *= experience_matrix[:, 0] / experience  # adjust for years of experience
    top_indices = similarity_scores.argsort()[::-1][:top_n]
    return df.iloc[top_indices]['jobtitle'].values

# Test the recommender system
print(get_similar_jobs('data analyst', 3))
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the data
data = pd.read_csv('/kaggle/input/jobs-on-naukricom/naukri_com-job_sample.csv')

# Replace NaN values with an empty string
data = data.replace(np.nan, '', regex=True)

# Select the columns for analysis
columns = ['skills', 'education']

# Combine the selected columns into a single feature column
data['features'] = data[columns].apply(lambda x: ' '.join(x), axis=1)

# Create a TF-IDF vectorizer to convert text to vectors
tfidf = TfidfVectorizer(stop_words='english')

# Fit and transform the feature column to obtain a matrix of vectorized text
tfidf_matrix = tfidf.fit_transform(data['features'])

# Compute the cosine similarity between all pairs of job postings
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Define a function to get the top recommendations based on cosine similarity
def get_recommendations(title, cosine_sim=cosine_sim, data=data, top=2):
    # Get the index of the job posting that matches the title
    idx = data[data['jobtitle']==title].index[0]

    # Get the similarity scores between the job posting and all other job postings
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the job postings based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the top recommendations
    recommendations = sim_scores[1:top+1]

    # Get the indices of the top recommendations
    indices = [rec[0] for rec in recommendations]

    # Return the top recommendations as a dataframe
    return data.iloc[indices]

# Test the recommendation system by getting the top 10 recommendations for a job title
recommendations = get_recommendations('Data Scientist')
print(recommendations)
```
